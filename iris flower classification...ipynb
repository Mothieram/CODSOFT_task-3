```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
iris_data= sns.load_dataset('iris')
iris_data.head()
```

```python
iris_data['species'],categories= pd.factorize(iris_data['species'])
iris_data.head()
```

```python
sns.scatterplot(data=iris_data ,x="sepal_length",y="sepal_width",hue="species")
```

```python
sns.scatterplot(data=iris_data ,x="petal_length",y="petal_width",hue="species")
```

```python
K_rng=range(1,11)
sse=[]
for k in K_rng:
  km=KMeans(n_clusters=k)
  km.fit(iris_data[['petal_length','petal_width']])
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('kreg')
plt.ylabel("sum of the squared error")
plt.plot(K_rng,sse)
```

```python
km=KMeans(n_clusters=3,random_state=0)
y_predicted=km.fit_predict(iris_data[['petal_length','petal_width']])
y_predicted
```

```python
iris_data['cluster']=y_predicted
iris_data.head(150)
```

```python
from sklearn.metrics import confusion_matrix
df=iris_data
cm=confusion_matrix(df.species,df.cluster)
cm
```

```python
true_labels=df.species
predicted_labels=df.cluster

cm= confusion_matrix(true_labels,predicted_labels)
class_labels=['verginica','versicolor','setose']
plt.imshow(cm,interpolation='nearest',cmap=plt.cm.Blues)
plt.title('iris confusion matrix')
plt.colorbar()
tick_marks=np.arange(len(class_labels))
plt.xticks(tick_marks,class_labels)
plt.yticks(tick_marks,class_labels)

for i in range(len(class_labels)):
  for j in range(len(class_labels)):
      plt.text(j,i,str(cm[i][j]),ha='center',va='center',color='white')


plt.xlabel('predicted label')
plt.ylabel('true label')
plt.show()
```
